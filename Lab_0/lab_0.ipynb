### Exercise 0.0
- Generate a simple sine wave signal with target frequency of 1 [Hz] and amplitude 3 [-]
- Plot the signal with proper axis label 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define time array
t = np.arange(0, 10.01, 0.01)

# Define frequency
f = 1  # Hz

# Define amplitude
A = 3

# Generate signal 
y = A * np.sin(2 * np.pi * f * t) 


# Plot the generated wave
plt.figure(figsize = (10, 5))
plt.plot(t, y)
plt.xlabel('Time [s]', fontsize = 16)
plt.ylabel('Amplitude [-]', fontsize = 16)
plt.title('Sine wave', fontsize = 20)
plt.grid()
```

### Exercise 0.1
- Generate white noise (uniform random numbers between -1 and 1) and superimpose it to the previously generated signal. 
- Plot the histogram of the generated noise (50 bins).
- Plot the signal with proper axis label 
- Compute mean, std and RMS and save values to a .txt file.
- Save the signal to a .npy file
- Reload the signal from .npy file

```python
# Generate noise
noise = np.random.uniform(-1,1,len(t))

# Generate signal with added noise
y_noise = y + noise

# Plot the histogram of the noise
plt.hist(noise, bins=50, density=True)
plt.title("Histogram of White Noise")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Plot the generated wave
plt.figure(figsize = (10, 5))
plt.plot(t, y_noise)
plt.xlabel('time [s]', fontsize = 16)
plt.ylabel('amplitude [-]', fontsize = 16)
plt.title('1 [Hz] sin wave with noise', fontsize = 20)
plt.grid()

# Compute mean, standard deviation and RMS of the signal
mean = np.mean(y_noise)
std = np.std(y_noise)
rms = np.sqrt(np.mean(y_noise**2))

# Save the mean, std, and RMS values to a text file
with open("signal_stats.txt", "w") as f:
    f.write("Mean: {}\nStandard Deviation: {}\nRMS: {}".format(mean, std, rms))

# Save the signal to a .npy file
np.save("signal.npy", y_noise)

# Reload the signal from the .npy file
reloaded_y_noise = np.load("signal.npy")



```

### Exercise 0.2
- Plot the signal+noise with proper axis label and highlight all the values greater than target threshold = 2

```python
# Define threshold for selecting points
threshold = 2

# Find indices of points above threshold
idx = y_noise > threshold

# Select y-values and corresponding time values above threshold
y_select = y_noise[idx]
t_select = t[idx]

# Plot the sine wave and selected points
plt.figure(figsize = (10, 5)) # Set figure size
plt.plot(t, y_noise) # Plot the sine wave with noise

# Label x-axis and y-axis, set font size
plt.xlabel('time [s]', fontsize = 16)
plt.ylabel('amplitude [-]', fontsize = 16)

# Add title and set font size
plt.title('1 [Hz] sin wave with noise', fontsize = 20)

# Plot selected points in red with marker 'o'
plt.plot(t_select, y_select, 'o', color = 'red')

# Add horizontal line at threshold in red
plt.axhline(y=threshold, color='red', linestyle='--')

# Add grid to plot
plt.grid()
```
